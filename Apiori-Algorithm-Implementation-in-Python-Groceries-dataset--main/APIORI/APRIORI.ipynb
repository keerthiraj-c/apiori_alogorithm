## Data Preprocessing

## Training the Apriori model on the dataset

```python
dataset = pd.read_csv('groceries.csv')
transactions = [] #creating empty list
```

```python
dataset
```

```python
dataset.shape
```

```python
for i in range(0, 344):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 19)])
```

```python
transactions
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.002, min_confidence = 0.1, min_lift = 2, min_length = 2, max_length = 2)
print(rules)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```python
results = list(rules) #to manipulate data easily using index
```

```python
results#[2][0][0])[0]
```

### Putting the results well organised into a Pandas DataFrame

```python
ans = [tuple(result[2][0][1])[0] for result in results] #to access list inside a list
ans
```

```python
supp    = [result[1] for result in results]
supp
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
